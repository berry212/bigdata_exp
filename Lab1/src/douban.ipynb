```python
import re, requests, os, wget
from bs4 import BeautifulSoup
```

```python
url = 'https://movie.douban.com/top250'

header = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/128.0.0.0 Safari/537.36 Edg/128.0.0.0'
}

outPath = '../out'
```

```python
resp = requests.get(url, headers=header)
resp
```

```python
print(resp.text)
```

```python
def parseContent_img(content, fp):
    soup = BeautifulSoup(content, 'html.parser')
    data = soup.findAll('img')
```

```python
content = resp.text
soup = BeautifulSoup(content, 'html.parser')
data = soup.findAll('img')
data
```

```python
for i in data:
    s = str(i)
    alt_match = re.search(r'alt="(.*?)"', s)
    src_match = re.search(r'src="(.*?)"', s)
    if alt_match and src_match:
        alt = alt_match.group(1)
        src = src_match.group(1)
        print(f"{alt}, {src}")
        wget.download(url=src, out=os.path.join(outPath, 'photo', alt + '.jpg'))
```

```python
def parseContent_title(content, fp):
    soup = BeautifulSoup(content, 'html.parser')
    data = soup.findAll('span', attrs={'class' : 'title'})
    for i in data:
        if '/' not in i.string:
            print(i.string, file=fp)
```

```python
def parseContent_img(content):
    soup = BeautifulSoup(content, 'html.parser')
    data = soup.findAll('img')

    for i in data:
        s = str(i)
        alt_match = re.search(r'alt="(.*?)"', s)
        src_match = re.search(r'src="(.*?)"', s)
        if alt_match and src_match:
            alt = alt_match.group(1)
            src = src_match.group(1)
            print(f"{alt}, {src}")
            wget.download(url=src, out=os.path.join(outPath, 'photo', alt + '.jpg'))
```

```python
def outTitle():
    fp = open(os.path.join(outPath, 'Top250.txt'), 'w+')
    for i in range(10):

        args = 'start=' + str(i*25)
        resp = requests.get(url + '?' + args, headers=header)
        content = resp.text

        parseContent_title(content, fp)
        parseContent_img(content)
    fp.close()


outTitle()
```
